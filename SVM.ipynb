### SVM as in homl

```python
import numpy as np
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
```

```python
iris = datasets.load_iris()
```

```python
X = iris['data'][:,2:]
y = (iris['target']== 2).astype(int)

lin_svc = Pipeline([
    ('standard', StandardScaler()),
    ('linsvc', LinearSVC(C=1, loss='hinge'))
])
lin_svc.fit(X,y)
#Other possibilities to train a linear support vector machine are SVC(kernel='linear') or SGDClassifier(loss='hinge')
```

```python
lin_svc.predict([[5.5,1.7]])
```

## Non-linear SVMs 
##### Data that is not separable using a linear SVM can be separated with a linear SVM by adding extra features such as PolynomialFeatures

```python
from sklearn.datasets import make_moons
from sklearn.preprocessing import PolynomialFeatures
from matplotlib import pyplot as plt
lin_svm = Pipeline([
    ('poly', PolynomialFeatures(degree=3)),
    ('standard', StandardScaler()),
    ('linearsvm', LinearSVC(loss='hinge', C=10, max_iter=10000))
])
X,y = make_moons(n_samples=100, noise=0.15)
lin_svm.fit(X,y)
```

```python
plt.scatter(X[:,0],X[:,1], c=y)
```
